```python
##FINDING OPTIMAL FEATURES FOR MODELLING USING RECURSIVE FEATURE ELIMINATION
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import StratifiedKFold
from sklearn.feature_selection import RFE
from sklearn.feature_selection import RFECV
```

```python
y=df1.CD4cat
X=df1.drop(['Unnamed: 0','CD4cat','Tanzania'],axis=1)
model = LogisticRegression(random_state=10)
rfe = RFE(estimator=model,step=1)
rfe.fit(X,y)
```

```python
print('optimal number of features: {}'.format(rfe.n_features_))
```

```python
selected_features = pd.DataFrame({'feature':list(X.columns),'Ranking':rfe.ranking_})
selected_features.sort_values(by='Ranking')
```

```python
##DEVELOPING A LOGISTIC REGRESSION WITH TOP 10 FEATURES
from sklearn.model_selection import train_test_split
X = df1[['pregstatusnw','LAGrecentorlonginfec','hadsexinterc','avoidpregnancy','marriedorlivedtogether','ARVsdetected','attendedschool','worksicklast3mon','enrolledschool','liveinhouse']]
y = df1.CD4cat
X_train, X_test, y_train, y_test =train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
##CHECKING ACCURACY OF THE MODEL
model.score(X_test,y_test)
```
